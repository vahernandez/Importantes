```python
dicc1 = {
    10:7,
    20:(1,2,3),
    30:['Control', 'Educación']
}

print(dicc1)
```

```python
# Acceder al string 'Educación'
# Esta en el identificador 30 y dentro del indentificador, esta en la lista posición 1
# Vamos de lo mas hacia afuera hacia lo mas hacia dentro

print(dicc1[30][1])
```

```python
datos = {
    'nombre': 'Sergio',
    'edad' : 25,
    'Genero': 'M'
}
```

```python
#Borrando el elemento "edad":
del datos['edad']
print(datos)
```

```python
#Imprimiendo solo los valores de cada llave:

print(datos.values())
```

```python
#Imprimiendo solo los índices:
print(datos.keys())
```

```python
#Para imprimir los dos:
print(datos.items())
```

```python
datos['edad'] = 25
print(datos)
# Iterando con for:
for i, j in datos.items():
    print(f'su {i} es {j}')
```
